```python
import pandas as pd
```

```python
df = pd.read_csv("HRGP_raw_both_strand.csv",index_col=0)
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")    
grid = sns.JointGrid(iris.petal_length, iris.petal_width, space=0, size=6, ratio=50)
grid.plot_joint(plt.scatter, color="g")
plt.plot([0, 4], [1.5, 0], linewidth=2)
```

```python
df.head()
```

```python
number = df.groupby('class').sum()
```

```python
number
```

```python
sel_cols = []
for c in number.columns:
    if number.at[0.0,c] <= 29 and number.at[1.0,c] >= 11:
        sel_cols.append(c)
len(sel_cols)
```

```python
number
```

```python
plot_df = number.T
plot_df.columns = ['Neg','Pos']
```

```python
plt.figure(figsize=(40,20))
grid = sns.JointGrid(plot_df.Pos, plot_df.Neg, space=0, size=6)
grid.plot_joint(plt.scatter, color="darkblue",alpha=0.4)
plt.plot([0, 11], [29, 11], linewidth=2,color="red")
plt.plot([10.5, 29.5], [13, 29.5], linewidth=2,color="red")
plt.xlim([-0.5,13.5])
plt.xticks(list(range(14)))


```

```python

f = plt.figure(figsize=(10,4))
grid = sns.JointGrid(plot_df.Pos, plot_df.Neg, space=0, size=6)
grid.plot_joint(plt.scatter, color="darkblue",alpha=0.4)
grid.fig.set_figwidth(10)
grid.fig.set_figheight(4)
plt.plot([10.5, 10.5], [-5, 29.5], linewidth=2,color="red")
plt.plot([10.5, 13.5], [29.5, 29.5], linewidth=2,color="red")
plt.xlim([-0.5,13.5])
plt.ylim([-5,132])
plt.xticks(list(range(14)))
plt.xlabel("Number occurrences in 13 pollen-specific genes")
plt.ylabel("Number occurrences in 132 non-pollen genes")

grid.savefig("Max.png", dpi=300)
```
